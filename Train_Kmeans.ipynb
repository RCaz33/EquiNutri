```python
import pandas as pd
data = pd.read_csv('CleanedOpenFoodFacts.csv', sep="\t",low_memory=False)
print("import done")
data.head()
```

```python
def grams_to_energy(data):
    return data['fat_100g'].values * 37, \
        data['carbohydrates_100g'].values * 17, \
            data['proteins_100g'].values * 17

# energi
data['NRJ_fat'] = data['fat_100g'].values * 37
data['NRJ_carb'] = data['carbohydrates_100g'].values * 17
data['NRJ_prot'] = data['proteins_100g'].values * 17
```

```python
# make dummies of important categories

# pnns_groups_1
pd.concat([data , data['pnns_groups_1'].str.get_dummies()], 
          axis = 1)

# nutrition_grade_fr
data.loc[data.nutrition_grade_fr.isnull(), 'nutrition_grade_fr'] = 'z'
pd.concat([data , data['nutrition_grade_fr'].str.get_dummies()], 
          axis = 1)
```
